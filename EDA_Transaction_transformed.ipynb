##### Visualizar distribucion de datos
##### Histograma de la columna total


```python
plt.figure(figsize=(10, 6))
sns.histplot(df_data['total'], kde=True)
plt.title('Distribucin de la Columna total')
plt.xlabel('total')
plt.ylabel('Frecuencia')
plt.show()
```

##### Boxplot de la Columna total

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_data['total'])
plt.title('Boxplot de la Columna total')
plt.xlabel('total')
plt.show()
```

##### Filtrar solo las columnas numricas

```python
df_numerico = df_data.select_dtypes(include=['float64', 'int64'])
```

##### Calcular la matriz de correlacin

```python
correlation_matrix = df_numerico.corr()
```

##### Visualizar la matriz de correlacin

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlacin')
plt.show()
```
